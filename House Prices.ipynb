```python
data = train.select_dtypes(include=[np.number]).interpolate().dropna()
```

```python
sum(data.isnull().sum() != 0)
```

```python
y = np.log(train.SalePrice)
X = data.drop(['SalePrice', 'Id'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)
```

```python
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
print "R^2 is: " + str(model.score(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
print "RMSE is: " + str(mean_squared_error(y_test, predictions))
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values, color='b')
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model')
plt.show()
```

```python
for i in range(-2, 3):
    alpha = 10**i
    rm = linear_model.Ridge(alpha=alpha)
    ridge_model = rm.fit(X_train, y_train)
    preds_ridge = ridge_model.predict(X_test)
    
    plt.scatter(preds_ridge, actual_values, color='b')
    plt.xlabel('Predicted Price')
    plt.ylabel('Actual Price')
    plt.title('Ridge Regularization with alpha=%s'% alpha)
    overlay = 'R^2 is: %s and RMSE is: %s'% (ridge_model.score(X_test, y_test), mean_squared_error(y_test, preds_ridge))
    plt.annotate(s=overlay, xy=(12.1, 10.6), size='x-large')
    plt.show()
```
