```python
def encode(x):
    return 1 if x == 'Partial' else 0
```

```python
train['enc_sale_condition'] = train.SaleCondition.apply(encode)
test['enc_sale_condition'] = test.SaleCondition.apply(encode)
```

```python
enc_sale_condition_pivot = train.pivot_table(index='enc_sale_condition', values='SalePrice', aggfunc=np.median)
enc_sale_condition_pivot.plot(kind='bar', color='blue')
plt.xlabel('Sale Condition')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```
